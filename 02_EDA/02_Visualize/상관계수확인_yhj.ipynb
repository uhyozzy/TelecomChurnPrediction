## **각 컬럼별 상관계수의 절대값이 0.3 이상인 경우**

```python
def find_high_correlation_columns(corr_matrix, threshold=0.3):
    """
    주어진 상관 관계 행렬에서 각 컬럼별로 상관 관계가 주어진 임계값(threshold) 이상인 컬럼을 출력합니다.

    매개변수:
    corr_matrix (pd.DataFrame): 상관 관계 행렬.
    threshold (float): 상관 관계의 절대값이 이 임계값 이상인 컬럼을 선택합니다.

    반환값:
    None (출력만 수행)
    """
    # 대각선 요소를 0으로 만듭니다.
    for i in range(len(corr_matrix.columns)):
        corr_matrix.iloc[i, i] = 0

    # 각 컬럼별로 상관 관계가 임계값 이상인 컬럼을 출력합니다.
    for column in corr_matrix.columns:
        correlated_columns = corr_matrix[column][corr_matrix[column] >= threshold].index.tolist()
        if correlated_columns:
            print(f"'{column}' 상관계수가 {threshold} 이상인 경우: {', '.join(correlated_columns)}\n")

# demonserv_corr를 사용하여 각 컬럼별로 0.3 이상의 상관 관계를 출력합니다.
find_high_correlation_columns(demonserv_corr, threshold=0.3)

```

## **각 컬럼별 상관계수의 절대값이 높은 5개**

```python
def find_top_correlation_columns(corr_matrix, num_top_columns=5):
    """
    주어진 상관 관계 행렬에서 각 컬럼별로 상관 관계가 가장 높은 상위 num_top_columns개의 컬럼을 출력합니다.

    매개변수:
    corr_matrix (pd.DataFrame): 상관 관계 행렬.
    num_top_columns (int): 출력할 상위 컬럼의 개수.

    반환값:
    None (출력만 수행)
    """
    # 대각선 요소를 0으로 만듭니다.
    for i in range(len(corr_matrix.columns)):
        corr_matrix.iloc[i, i] = 0

    # 각 컬럼별로 상관 관계가 가장 높은 상위 num_top_columns개의 컬럼을 출력합니다.
    for column in corr_matrix.columns:
        top_correlated_columns = corr_matrix[column].nlargest(num_top_columns).index.tolist()
        print(f"'{column}'과(와) 상관계수가 높은 컬럼: {', '.join(top_correlated_columns)}\n")

# demonserv_corr를 사용하여 각 컬럼별로 상관 관계가 가장 높은 상위 5개의 컬럼을 출력합니다.
find_top_correlation_columns(demonserv_corr, num_top_columns=5)
```
