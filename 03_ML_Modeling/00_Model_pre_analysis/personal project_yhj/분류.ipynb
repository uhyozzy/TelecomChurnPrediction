## **1-1. 데이터타입 조정**
- Membership 카테고리 컬럼을 5단계로 나누어 변경한다.
  - Offer A: 1, Offer B: 2, Offer C: 3, Offer D: 4, Offer E: 5, None: 0
  - 단, 0은 멤버쉽이 아님을 의미하며 'A→E'는 멤버쉽 정도의 '낮음→높음'을 의미한다.
- Contract 컬럼의 계약 유형도 개월수 데이터로 변경해준다
  - Month-to-month : 1, One year : 12, Two year : 24

```python
# Offer A: 1, Offer B: 2, Offer C: 3, Offer D: 4, Offer E: 5, None: 0 로 변경

df.loc[df["Membership"]=='Offer A', "Membership"] = 1
df.loc[df["Membership"]=='Offer B', "Membership"] = 2
df.loc[df["Membership"]=='Offer C', "Membership"] = 3
df.loc[df["Membership"]=='Offer D', "Membership"] = 4
df.loc[df["Membership"]=='Offer E', "Membership"] = 5
df.loc[df["Membership"]=='None', "Membership"] = 0

df["Membership"] = df["Membership"].astype('int64')
```

```python
# Month-to-month:1, One year:12, Two year:24 로 변경
df.loc[df["Contract"]=='Month-to-Month', "Contract"] = 1
df.loc[df["Contract"]=='One Year', "Contract"] = 12
df.loc[df["Contract"]=='Two Year', "Contract"] = 24

df["Contract"] = df["Contract"].astype('int64')
```

## **1-2. 가중치 추가**

**가중치 추가 전 상관관계**

```python
# 'New Churn Score'와 다른 열들 간의 상관계수 계산
correlation0 = data.corr()['Churn Score']
correlation0
```


-  EDA 결과, 각 features 별 Churn Score와의 상관관계는 다음과 같았다.
  - Number of Dependents    : 부양가족이 많을수록 이탈 확률이 낮음
  - Tenure in Months        : 이용개월수가 많을수록 이탈 확률이 낮음
  - Tech Use                : 기술 서비스 이용 횟수가 많을수록 이탈 확률이 낮음
  - Contract	           : 계약의 규모가 높을수록 이탈 확률이 낮음
  - Satisfaction Score    : 상담 만족도가 높을수록 이탈 확률이 낮음

- 임의로 추가 및 보정할 컬럼
  - Age Group               : 나이가 많을 수록 이탈 확률이 낮음
  - Combined Product        : 결합 서비스를 이용할수록 이탈 확률이 낮음
  - Membership              : 멤버십 단계가 높을수록 이탈 확률이 낮음

```python
# 상관도를 목적에 맞게 보정할 수 있도록 Churn Score의 표준편차(21)를 가감함

# 기본적으로 'Churn Score' 값을 복사
data['New Churn Score'] = data['Churn Score']

# 1단계: Age Group 값이 20 또는 30인 경우 'Churn Score' 값에 1 더하기
data.loc[(data['Age Group'] == 20) | (data['Age Group'] == 30), 'New Churn Score'] += 21

# 2단계: Combined Product 값이 1인 경우 'Churn Score' 값에 1 빼기
data.loc[data['Combined Product'] == 1, 'New Churn Score'] -= 21

# 3단계: Membership 값이 1인 경우 'Churn Score' 값에 1 빼기
data.loc[data['Membership'] == 1, 'New Churn Score'] -= 5
data.loc[data['Membership'] == 2, 'New Churn Score'] -= 10
data.loc[data['Membership'] == 3, 'New Churn Score'] -= 15
data.loc[data['Membership'] == 4, 'New Churn Score'] -= 20
data.loc[data['Membership'] == 5, 'New Churn Score'] -= 21
```

```python
# 'New Churn Score'와 다른 열들 간의 상관계수 계산
correlation = data.corr()['New Churn Score']
correlation
```

**모델링용 최종 데이터셋 생성**

```python
# 모델링에 필요없는 컬럼제거 후 새로운 데이터 프레임 생성
data_set = data.drop(['Age Group','Churn Score'], axis=1)
data_set
```

```python
# 'New Churn Score'의 값이
# 40미만이라면 'Score Group'컬럼에 '안정 그룹'
# 40이상 60미만이라면 'Score Group'컬럼에 '잠재적 이탈 그룹'
# 60이상이라면 'Score Group'컬럼에 '고위험 이탈 그룹'

data_set['Score Group'] = pd.cut(data_set['New Churn Score'], bins=[0, 40, 60, float('inf')],
                                 labels=['안정 그룹', '잠재적 이탈 그룹', '고위험 이탈 그룹'])

data_set
```

```python
data_set.describe()
```

## **1-3. Encoding (MinMaxScaler)**

### **1-3-1. Features Encoding**

```python
from sklearn.preprocessing import MinMaxScaler

num_cols = ['Age', 'Number of Dependents', 'Satisfaction Score',
            'Tenure in Months', 'Tech Use', 'Combined Product',
            'Membership', 'Contract']

scaler = MinMaxScaler()
data_set[num_cols] = scaler.fit_transform(data_set[num_cols])
```

```python
data_set.head()
```

## **1-4. VIF계산 - 다중공선성 점검**
- 독립변수(X) 간에 상관관계가 존재하는 것(회귀식이 오류를 범할 수 있음)
  - VIF(분산팽창지수) 10 이상이면 다중공선성 존재
- 다중공선성을 야기하는 features는 없는 것으로 판단됨

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['VIF Factor'] = [variance_inflation_factor(
    data_set[num_cols].values, i) for i in range(data_set[num_cols].shape[1])]
vif['features'] = data_set[num_cols].columns
print(vif)
```

## **1-5. 학습/검증 데이터 생성**

```python
from sklearn.model_selection import train_test_split

# Label 데이터셋 추출
label_column = 'Score Group'
y = data_set[label_column]

# Feature 데이터셋 추출
x = data_set[num_cols]

# Feature 데이터셋과 Label 데이터셋을 train과 test 세트로 분할
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

**Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder

# LabelEncoder 객체 생성
label_encoder = LabelEncoder()

# 레이블 데이터에 대한 인코딩 수행
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

### **2-2-3. LightGBM**

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# LabelEncoder 객체 생성
label_encoder = LabelEncoder()

# 레이블 데이터에 대한 인코딩 수행
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# LightGBM 데이터셋 생성
train_data = lgb.Dataset(x_train, label=y_train_encoded)

# LightGBM 모델 설정
params = {
    'objective': 'multiclass',  # 다중 클래스 분류 문제
    'num_class': len(label_encoder.classes_),  # 클래스 수
    'boosting_type': 'gbdt',  # 그래디언트 부스팅 타입
    'metric': 'multi_logloss'  # 손실 함수
}

# LightGBM 모델 학습
num_round = 100  # 학습 라운드 수
bst = lgb.train(params, train_data, num_round)

# 테스트 데이터 예측
y_pred_encoded = bst.predict(x_test)
y_pred = [list(x).index(max(x)) for x in y_pred_encoded]  # 확률을 클래스로 변환

# 정확도 계산
accuracy = accuracy_score(y_test_encoded, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```
