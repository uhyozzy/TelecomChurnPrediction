## **(1) 통신기능**

```python
# 데이터프레임에 있는 컬럼을 사용하여 산점도 생성
fig = px.scatter(use_df, x="Avg Monthly Long Distance Charges", y="Churn Score")

fig.update_traces(showlegend=False) # 범례 생략
fig.show() # 그래프 표시
```

```python
ml_use_df = use_df.groupby('Multiple Lines')['Churn Score'].mean().round(2)
fig = px.bar( x = ml_use_df.index , y = ml_use_df.values,
             color = ml_use_df.index, text= ml_use_df.values )

fig.update_traces(
	textposition= "outside",
	textfont_size= 15
)

fig.update_layout(
	title = dict( text = "<b>Multiple Lines여부에 따른 Churn Score 평균</b>", font_size =20),
	xaxis = dict( title = "Multiple Lines"),
	yaxis = dict( title = "Churn Score 평균"),
	legend_title_text = "Multiple Lines", template = "plotly_white",
  width=600, height=500)


fig.show()
```

```python
ml_use_df = use_df.groupby('Internet Type')['Churn Score'].mean().round(2)
fig = px.bar( x = ml_use_df.index , y = ml_use_df.values,
             color = ml_use_df.index, text= ml_use_df.values )

fig.update_traces(
	textposition= "outside",
	textfont_size= 15
)

fig.update_layout(
	title = dict( text = "<b>Internet Type 사용 여부에 따른 Churn Score 평균</b>", font_size =20),
	xaxis = dict( title = "Internet Type"),
	yaxis = dict( title = "Churn Score 평균"),
	legend_title_text = "Internet Type", template = "plotly_white",
  width=600, height=500)


fig.show()
```
