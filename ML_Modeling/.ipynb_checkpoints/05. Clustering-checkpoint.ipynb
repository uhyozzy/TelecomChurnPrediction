# 데이터 로드

```python
%config Completer.use_jedi = False
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta

%matplotlib inline
```

```python
df = pd.read_excel("total.xlsx")
print(df.shape)
df.head()
```

```python
# Customer ID & 선정한 컬럼 & 'Churn Label', 'Churn Score'
selected_cols = ['Customer ID','Dependents', 'Under 30', 'Monthly Charge', 'Total Revenue', 'Tenure in Months', 'Internet Service', 
                 'Device Protection Plan', 'Unlimited Data', 'Streaming Movies', 'Churn Label', 'Churn Score']

new_df = df[selected_cols]

new_df
```

# label 값 맞추기

## Yes, No -> 1, 0

```python
target_cols = ['Under 30','Dependents', 'Internet Service','Device Protection Plan','Streaming Movies', 'Unlimited Data','Churn Label']

for col in target_cols:
    new_df[col] = new_df[col].map({'Yes': 1, 'No': 0})
    

new_df.head()
```

# 표준화

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

scaled_df = new_df.copy()

# 또는 MinMaxScaler를 사용한 정규화
scaler = MinMaxScaler()
scaled_df[['Monthly Charge', 'Total Revenue']] = scaler.fit_transform(scaled_df[['Monthly Charge', 'Total Revenue']])
```

```python
scaled_df
```

# 최적의 cluster 개수 찾기

## Elbow method(엘보우 기법)

```python
import matplotlib.pyplot as plt

# 문자형인 Customer ID는 제외 & 독립변수만 진행
clustering_df = scaled_df.iloc[:, 1:-2]

# 클러스터링을 진행할 클러스터 개수 범위 설정 (예: 2에서 10까지)
cluster_range = range(2, 11)
sse = []  # 클러스터 내 제곱합(SSE)을 저장할 리스트

for num_clusters in cluster_range:
    kmeans = KMeans(n_clusters=num_clusters, random_state=0)
    kmeans.fit(clustering_df)
    sse.append(kmeans.inertia_)  # 클러스터 내 제곱합(SSE) 계산 및 저장

# SSE 값을 그래프로 시각화하여 엘보우 지점 찾기
plt.figure(figsize=(8, 6))
plt.plot(cluster_range, sse, marker='o')
plt.title('Elbow Method for Optimal Cluster Number')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.grid()
plt.show()
```

SSE 값의 감소가 둔화되는 지점이 최적의 클러스터 개수로 선택 => 3개

## Silhouette Index(실루엣 인덱스)
clustering 기법에서 실질적으로 가장 많이 쓰이는 타당성(평가) 지표이다.

개별객체마다 silhouette index 값을 구할수 있으며, 이에대한 평균값을 사용한다.

모든값이 0.5 이상을 가지면 유의하다고 판단

```python
from sklearn.metrics import silhouette_score

# 문자형인 Customer ID는 제외 & 독립변수만 진행
clustering_df = scaled_df.iloc[:, 1:-2]

for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=0)
    cluster_labels = kmeans.fit_predict(clustering_df)
    silhouette_avg = silhouette_score(clustering_df, cluster_labels)
    print(f'cluster : {k} // silhouette index {silhouette_avg}')
```

cluster 2개에서 가장 높은 index 값을 가졌으므로 2가 가장 유의한 cluster 개수

```python
!pip install yellowbrick
import matplotlib.pyplot as plt
from yellowbrick.cluster import SilhouetteVisualizer

# 군집 수 범위 설정
cluster_range = range(2, 6)

# 하나의 plot에 여러 실루엣 그래프 그리기
fig, axes = plt.subplots(1, len(cluster_range), figsize=(15, 5))

for i, k in enumerate(cluster_range):
    kmeans = KMeans(n_clusters=k, random_state=0, init='random')
    visualizer = SilhouetteVisualizer(kmeans, colors='yellowbrick', ax=axes[i])
    visualizer.fit(clustering_df)
    axes[i].set_title(f'Cluster {k}')

plt.show()
```

## 최적의 cluster 개수
엘보우 기법 : 3개일 때 최적</br>
실루엣 인덱스 방법 : 2개일 때 최적

=> 엘보우 기법으로 선택

# Cluster

```python
# 문자형인 Customer ID는 제외 & 독립변수만 진행
clustering_df = scaled_df.iloc[:, 1:-2]

# scaled_df 데이터를 넣어 클러스터링
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit(clustering_df)

# 클러스터링 변수인 clusters 값을 원본 데이터인 'new_df'내에 넣기 
new_df['cluster'] = clusters.labels_
new_df.head()
```

## 클러스터별 개수

```python
#cluster를 기준으로 데이터 개수 세기
new_df.groupby('cluster').count()
```

## 클러스터별 특징

```python
# 플롯
plt.scatter(new_df['Dependents'], new_df['Under 30'], c=new_df['cluster'])
plt.xlabel("Dependents")
plt.ylabel("Under 30")

    
plt.show()
```
