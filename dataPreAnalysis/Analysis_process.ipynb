```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
labeling_data = pd.read_csv("./result_data/labeling_data.csv", header=0, index_col=0, encoding='utf-8')
```

## Pandas의 corr를 활용한 단순 상관계수 확인

```python
labeling_data_modi = labeling_data.copy()

labeling_data_modi.drop('Churn Reason_label',axis=1, inplace=True)
labeling_data_modi.drop('Churn Category_label',axis=1, inplace=True)
labeling_data_modi.corr()
```

```python
# 해지 여부 컬럼만 가져와서 상관계수 내림차순으로 정리
pd.DataFrame(labeling_data_modi.corr()['Churn Label_label']).sort_values('Churn Label_label', ascending=False)
```

### Result

- Customer Status는 고객의 현 상태를 나타내는 컬럼이다 보니, 상관 계수가 높은것이 당연함
- (추가)범주형 컬럼들을 가지고 그냥 계산하였기에 그냥 참고용으로 사용. 실제는 카이제곱을 사용해봐야 나올 것으로 예상됨

```python
# 일반 요금이 아닌, 요금제 형식으로 구분하여 확인
# 낮은 요금제를 사용하는 유저는, 이탈률이 높을까? 하는 의구심
# 균일하게 요금제를 나누는 방식보다, 특정 구간에 맞추어서 나눈 후에 적용
TC_df = pd.DataFrame(labeling_data[['Total Charges', 'Churn Label_label', 'Customer Status']])
bins_list = [0,100,250,500,1000,2000,4000,8000,9999]
labels_list=['lv0', 'lv1', 'lv2', 'lv3', 'lv4', 'lv5', 'lv6', 'lv7']
# boundary_range = pd.cut(labeling_data['Total Charges'], bins=8)
boundary_cut = pd.cut(labeling_data['Total Charges'], bins=bins_list, labels=labels_list)
# TC_df['Rate_Range'] = boundary_range
TC_df['Rate_Plan'] = boundary_cut
TC_df

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(labels_list)
TC_df['RP_label'] = pd.DataFrame(le.transform(TC_df['Rate_Plan']))
TC_df

TC_df.corr()
```
