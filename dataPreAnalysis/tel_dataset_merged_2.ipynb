```python
import numpy as np
import pandas as pd
```

```python
# 개별적으로 불러오기
tccsv = pd.read_csv('./data/Telco_customer_churn_services.csv', header=0, encoding='utf-8')
tccst = pd.read_csv('./data/Telco_customer_churn_status.csv', header=0, encoding='utf-8')
tccd = pd.read_csv('./data/Telco_customer_churn_demographics.csv', header=0, encoding='utf-8')
tccl = pd.read_csv('./data/Telco_customer_churn_location.csv', header=0, encoding='utf-8')
```

```python
# 동적 변수 사용하여 한번에 불러오기(참조, 사용하지 않음)
variable_name = ['tccsv2', 'tccst2', 'tccd2','tccl2']
file_names = ['Telco_customer_churn_services', 'Telco_customer_churn_status', 'Telco_customer_churn_demographics', 'Telco_customer_churn_location']

for i in range(len(file_names)):
    globals()[variable_name[i]] = pd.read_csv('./data/'+file_names[i]+'.csv', header=0, encoding='utf-8')
```

```python
# 공통 컬럼을 기준으로 하여, outer merge

merged_1 = pd.merge(tccsv, tccst, how='outer', on=['Customer ID', 'Count', 'Quarter'])
merged_2 = pd.merge(tccd, tccl, how='outer', on=['Customer ID', 'Count'])

merged_end = pd.merge(merged_2, merged_1, how='outer', on=['Customer ID', 'Count'])
```

```python
# Save as
merged_end.to_csv("./result_data/merged_date_2.csv")
```
