```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder
```

```python
# Data read
target_data = pd.read_csv("./result_data/merged_date_2.csv", header=0, index_col=0, encoding='utf-8')
```

```python
# Data check
target_data.head()
```

```python
# target_data[target_data.columns[0]].dtype == 'object'
```

```python
# len(target_data['Location ID'].unique())
```

```python
# categorical data check + dict create
categorical_column_list = []
for i in range(len(target_data.columns)):
    if target_data[target_data.columns[i]].dtype == 'object':
        if len(target_data[target_data.columns[i]].unique()) <= 100 and len(target_data[target_data.columns[i]].unique()) != 1:
        # if len(target_data[target_data.columns[i]].unique()) != len(target_data):
            categorical_column_list.append(target_data.columns[i])

categorical_column_list
```

```python
categorical_values_list = set()
for col_name in categorical_column_list:
    values = tuple(target_data[col_name].unique())
    categorical_values_list.add(values)

list(categorical_values_list)
```

```python
target_data['Gender'].dtype == 'object'
```

```python
encoder = LabelEncoder()
idx = 0

for df_col_name in target_data.columns:
    if target_data[df_col_name].dtype == 'object':
        if len(target_data[df_col_name].unique()) == len(target_data) or len(target_data[df_col_name].unique()) == 1:
            continue
        for j in range(len(list(categorical_values_list))):
            if tuple(target_data[df_col_name].unique()) == list(categorical_values_list)[j]:
                encoder.fit(list(categorical_values_list)[j])
                encode_result = encoder.transform(target_data[df_col_name])
                
                if idx == 0:
                    column_name = [df_col_name+"_label"]
                    result_df = pd.DataFrame(encode_result, columns=column_name)
                    idx += 1
                else:
                    column_name = [df_col_name+"_label"]
                    adding_df = pd.DataFrame(encode_result, columns=column_name)
                    result_df = pd.concat([result_df, adding_df], axis=1)
                    idx += 1

result_df
```

```python
final_data = pd.concat([target_data, result_df], axis=1)
```

```python
final_data.to_csv("./result_data/labeling_data.csv", encoding='utf-8')
```
