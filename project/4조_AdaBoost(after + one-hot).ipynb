## Modeling 

### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import plot_confusion_matrix
from sklearn import metrics
from sklearn.feature_selection import RFE
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import learning_curve
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
# RandomForestClassifier 모델 생성
# n_estimators=500: 결정 트리 갯수
# n_jobs = -1 : 모든 가능한 CPU 코어를 사용하여 모델 훈련을 병렬로 수행하도록 설정
# oob_score = True : 훈련 중에 선택되지 않은 데이터로 모델을 검증하는 데 도움
# max_features = "auto": 각 결정 트리가 분할할 때 고려할 최대 특성 개수의 제곱근 사용
# max_leaf_nodes = 30 : 각 결정 트리에서 생성될 수 있는 최대 리프 노드(말단 노드)의 개수를 제한
model_rf = RandomForestClassifier(n_estimators=500, random_state=30, n_jobs = -1,
                           oob_score = True, max_features = "auto", max_leaf_nodes = 30)

# 모델 학습
model_rf.fit(X_train, y_train)

# 테스트 데이터에 대한 클래스 확률 예측
probs_rf = model_rf.predict_proba(X_test)
 
# 클래스 1에 대한 확률을 사용하여 예측
threshold = 0.5  # 임계값 설정
pred_y = (probs_rf[:, 1] >= threshold).astype(int)

print (metrics.accuracy_score(y_test, pred_y))
```

```python
print(classification_report(y_test, pred_y))
plot_confusion_matrix(model_rf, X_test, y_test) 
```

#### RandomForest 결과로 Feature importance 파악하기

#### 훈련 세트와 테스트 세트 간의 성능 비교

```python
# 훈련 세트와 테스트 세트 간의 성능 비교
model_rf = RandomForestClassifier(n_estimators=500, random_state=30, n_jobs = -1,
                           oob_score = True, max_features = "auto", max_leaf_nodes = 30)
model_rf.fit(X_train, y_train)

# 훈련 세트에서의 정확도 계산
train_accuracy = model_rf.score(X_train, y_train)

# 테스트 세트에서의 정확도 계산
test_accuracy = model_rf.score(X_test, y_test)

# 결과 출력
print("훈련 세트 정확도:", train_accuracy)
print("테스트 세트 정확도:", test_accuracy)
```

#### Cross Validation Score for Random Forest Classifier

```python
scores = cross_val_score(model_rf, X, y, cv=3, scoring='f1')
print(scores)
print(f"Random Forest F1_score: {scores.mean() * 100:.2f}% +/- ({scores.std() * 100:.2f})")
```

#### 학습 곡선 (Learning Curve)

```python
# 학습 곡선을 그릴 모델과 데이터를 준비합니다
# model: 학습할 모델 (예: Logistic Regression, Decision Tree, 등)
# X: 특성 데이터 (Feature data)
# y: 목표 변수 (Target variable)

# learning_curve 함수를 사용하여 학습 곡선 데이터를 생성합니다
train_sizes, train_scores, test_scores = learning_curve(model_rf, X, y, train_sizes=np.linspace(0.1, 1.0, 10), cv=5)

# train_sizes: 학습 데이터 크기의 배열
# train_scores: 훈련 데이터에서의 성능 점수
# test_scores: 테스트 데이터에서의 성능 점수

# 평균과 표준 편차를 계산합니다
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

# 학습 곡선을 그립니다
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, label='Training Score', color='blue', marker='o')
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color='blue', alpha=0.25)
plt.plot(train_sizes, test_mean, label='Cross-Validation Score', color='green', marker='s')
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, color='green', alpha=0.25)

plt.title('Learning Curve')
plt.xlabel('Training Examples')
plt.ylabel('Score')
plt.legend(loc='best')
plt.grid()
plt.show()
```

### Hyperparameter Optimization

```python
lr_params = {"C": [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
knn_params = {"n_neighbors": [3, 5, 7]}
dt_params = {"max_depth": [3, 5, 7]}
rf_params = {"n_estimators": [100, 300, 500]}
ada_params = {"base_estimator": [DecisionTreeClassifier(max_depth=3), DecisionTreeClassifier(max_depth=5)],
    "learning_rate": [0.01, 0.1]}
```

```python
classifiers = [
    ("LR", LogisticRegression(), lr_params),
    ("KNN", KNeighborsClassifier(), knn_params),
    ("dt", DecisionTreeClassifier(), dt_params),
    ("RF", RandomForestClassifier(), rf_params),
    ("ADA", AdaBoostClassifier(), ada_params)
]
```

```python
def hyperparameter_optimization(X, y, cv=5, scoring="accuracy"):
    print("Hyperparameter Optimization")
    best_models = {}
    for name, classifier, params in classifiers:
        print(f"########## {name} #########")
        gs_best = GridSearchCV(classifier, params, cv=cv, n_jobs=-1, verbose=False).fit(X, y)
        final_model = classifier.set_params(**gs_best.best_params_)

        cv_results = cross_validate(final_model, X, y, cv=cv, scoring=scoring)
        print(f"{scoring} (After) : {round(cv_results['test_score'].mean(), 4)}")
        print(f"{name} best params: {gs_best.best_params_}", end="\n\n")
        best_models[name] = final_model
    return best_models
```

```python
best_models = hyperparameter_optimization(X, y)
```
