## Random Forest Classifier -> Overfitting

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import plot_confusion_matrix
from sklearn import metrics
```

```python
# RandomForestClassifier 모델 생성
# n_estimators=500: 결정 트리 갯수
# n_jobs = -1 : 모든 가능한 CPU 코어를 사용하여 모델 훈련을 병렬로 수행하도록 설정
# oob_score = True : 훈련 중에 선택되지 않은 데이터로 모델을 검증하는 데 도움
# max_features = "auto": 각 결정 트리가 분할할 때 고려할 최대 특성 개수의 제곱근 사용
# max_leaf_nodes = 30 : 각 결정 트리에서 생성될 수 있는 최대 리프 노드(말단 노드)의 개수를 제한
rf = RandomForestClassifier(n_estimators=500, random_state=30, n_jobs = -1,
                           oob_score = True, max_features = "auto", max_leaf_nodes = 30)

# 모델 학습
rf.fit(X_train, y_train)

# 학습된 모델을 사용하여 test 데이터에 대한 예측 수행 
pred_y = rf.predict(X_test)
print (metrics.accuracy_score(y_test, pred_y))
```

```python
print(classification_report(y_test, pred_y))
plot_confusion_matrix(rf, X_test, y_test) 
```
