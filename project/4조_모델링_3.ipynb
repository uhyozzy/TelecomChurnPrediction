## Modeling

### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import plot_confusion_matrix
from sklearn import metrics
from sklearn.feature_selection import RFE
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import learning_curve
```

```python
# RandomForestClassifier 모델 생성
# n_estimators=500: 결정 트리 갯수
# n_jobs = -1 : 모든 가능한 CPU 코어를 사용하여 모델 훈련을 병렬로 수행하도록 설정
# oob_score = True : 훈련 중에 선택되지 않은 데이터로 모델을 검증하는 데 도움
# max_features = "auto": 각 결정 트리가 분할할 때 고려할 최대 특성 개수의 제곱근 사용
# max_leaf_nodes = 30 : 각 결정 트리에서 생성될 수 있는 최대 리프 노드(말단 노드)의 개수를 제한
model_rf = RandomForestClassifier(n_estimators=500, random_state=30, n_jobs = -1,
                           oob_score = True, max_features = "auto", max_leaf_nodes = 30)

# 모델 학습
model_rf.fit(X_train, y_train)

# 테스트 데이터에 대한 클래스 확률 예측
probs_rf = model_rf.predict_proba(X_test)
 
# 클래스 1에 대한 확률을 사용하여 예측
threshold = 0.5  # 임계값 설정
pred_y = (probs_rf[:, 1] >= threshold).astype(int)

print (metrics.accuracy_score(y_test, pred_y))
```

```python
print(classification_report(y_test, pred_y))
plot_confusion_matrix(model_rf, X_test, y_test) 
```

#### RandomForest 결과로 Feature importance 파악하기

#### 클래스 레이블 예측을 수행
    * 클래스가 리프 노드에 도달한 비율을 계산

```python
# Decision Tree 모델 생성
model_dt_1 = DecisionTreeClassifier(random_state=42, max_depth=8, min_samples_split=10)

# 모델 학습
model_dt_1.fit(X_train, y_train)

# 테스트 데이터에 대한 클래스 확률 계산
leaf_indices = model_dt_1.apply(X_test)  # 각 샘플이 도달한 리프 노드의 인덱스를 가져옵니다.
n_classes = model_dt_1.n_classes_  # 클래스의 개수를 가져옵니다.

# 클래스 확률 초기화
class_probabilities = np.zeros((len(X_test), n_classes))

# 각 리프 노드에서 클래스 확률 계산
for i, leaf_index in enumerate(leaf_indices):
    leaf_values = model_dt_1.tree_.value[leaf_index]  # 리프 노드의 클래스 값 (레이블별 샘플 수)을 가져옵니다.
    class_probabilities[i] = leaf_values / leaf_values.sum()  # 클래스 확률을 계산합니다.

# 클래스 1에 대한 확률을 사용하여 예측
threshold = 0.5  # 임계값 설정
pred_y = (class_probabilities[:, 1] >= threshold).astype(int)

# 정확도 계산
accuracy_dt = accuracy_score(y_test, pred_y)
print("Decision Tree 정확도:", accuracy_dt)
```
