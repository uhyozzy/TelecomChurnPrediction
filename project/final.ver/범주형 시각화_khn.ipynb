```python
# Senior Citizen 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Senior Citizen'] = df['Senior Citizen'].map({'Yes': 1, 'No': 0})
```

```python
# Married 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Married'] = df['Married'].map({'Yes': 1, 'No': 0})
```

```python
# Dependents 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Dependents'] = df['Dependents'].map({'Yes': 1, 'No': 0})
```

```python
# Referred a Friend 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Referred a Friend'] = df['Referred a Friend'].map({'Yes': 1, 'No': 0})
```

```python
# Phone Service 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Phone Service'] = df['Phone Service'].map({'Yes': 1, 'No': 0})
```

```python
# Multiple Lines 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Multiple Lines'] = df['Multiple Lines'].map({'Yes': 1, 'No': 0})
```

```python
# Internet Service 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Internet Service'] = df['Internet Service'].map({'Yes': 1, 'No': 0})
```

```python
# Online Security 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Online Security'] = df['Online Security'].map({'Yes': 1, 'No': 0})
```

```python
# Online Backup 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Online Backup'] = df['Online Backup'].map({'Yes': 1, 'No': 0})
```

```python
# Device Protection Plan 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Device Protection Plan'] = df['Device Protection Plan'].map({'Yes': 1, 'No': 0})
```

```python
# Premium Tech Support 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Premium Tech Support'] = df['Premium Tech Support'].map({'Yes': 1, 'No': 0})
```

```python
# Streaming TV 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Streaming TV'] = df['Streaming TV'].map({'Yes': 1, 'No': 0})
```

```python
# Streaming Movies 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Streaming Movies'] = df['Streaming Movies'].map({'Yes': 1, 'No': 0})
```

```python
# Streaming Music 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Streaming Music'] = df['Streaming Music'].map({'Yes': 1, 'No': 0})
```

```python
# Unlimited Data 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Unlimited Data'] = df['Unlimited Data'].map({'Yes': 1, 'No': 0})
```

```python
# Paperless Billing 컬럼 값을 Yes을 1로, No 를 0으로 변경
df['Paperless Billing'] = df['Paperless Billing'].map({'Yes': 1, 'No': 0})
```

```python
# Customer Status 컬럼 값을 Stayed 를 0, Churned 을 1, Joined 을 2로 변환
# Churn : 1 의 경우 Churn Value 컬럼의 Yes:1 과 맞춤
df['Customer Status'] = df['Customer Status'].map({'Churned': 1, 'Stayed': 0, 'Joined': 2})
```

```python
# Churn Category 컬럼 값을 카테고리 별 숫자로 변환 (0~4)
df['Churn Category'] = df['Churn Category'].map({'Competitor': 1, 'Attitude': 2, 'Dissatisfaction': 3, 'Price': 4, 'Other': 5})
```

```python
# Offer 컬럼 값을 카테고리 별 숫자로 변환 (0~4)
df['Offer'] = df['Offer'].map({'Offer A': 1, 'Offer B': 2, 'Offer C': 3, 'Offer D': 4, 'Offer E': 5})
```

```python
# Internet Type 컬럼 값을 숫자로 변환 (0~2)
df['Internet Type'] = df['Internet Type'].map({'Fiber Optic': 1, 'DSL': 2, 'Cable': 3})
```

```python
# Contract 컬럼 값을 숫자로 변환 (0~2)
df['Contract'] = df['Contract'].map({'Month-to-Month': 0, 'One Year': 1, 'Two Year': 2})
```

```python
# Payment Method 컬럼 값을 숫자로 변환 (0~2)
df['Payment Method'] = df['Payment Method'].map({'Bank Withdrawal': 0, 'Credit Card': 1, 'Mailed Check': 2})
```

### 결측치 None으로 채워넣기

```python
# 'Churn Category'와 'Churn Reason' 컬럼의 결측치를 "None"으로 대체
df['Churn Category'].fillna(0, inplace=True)
df['Churn Reason'].fillna("None", inplace=True)
```

```python
# 'Offer'와 'Internet Type' 컬럼의 결측치를 "None"으로 대체
df['Offer'].fillna(0, inplace=True)
df['Internet Type'].fillna(0, inplace=True)
```

## CLTV 고객 분류별 시각화
    
    * 2000-4000: 저가치 고객
    * 4001-6000: 잠재 이탈율이 큰 고객으로 우리가 눈여겨봐야할 고객층
    * 6000 이상: 고가치 고객

```python
# csv 파일 불러오기
df = pd.read_csv('./data/Churn.csv')
```

```python
df.head()
```

```python
# CLTV 컬럼에서 점수대를 나눔
df['CLTV Group'] = pd.cut(df['CLTV'], bins=[2000, 4000, 6000, 7000], labels=['2000-4000', '4001-6000', '6000+'])
```

```python
df['Total Revenue Group'] = pd.cut(df['Total Revenue'], bins=[0, 2000, 6000, 10000, float('inf')], labels=['0-2000', '2001-6000', '6001-10000', '10000+'])
```

```python
plt.figure(figsize=(12,5),dpi=200)
sns.countplot(data=df, x='Total Revenue Group', hue='CLTV Group', palette='Set2')
```

* CLTV 4001-6000 대에 총 수익을 많이 가져다주는 유저들이 몰려있으며, 6000이 넘어가면서 CLTV 가 낮은 점수의 사용자가 거의 없음.

### Tenure 그룹별 CLTV

* 5년 이상
* 2년 이상 ~ 5년 미만
* 2년 미만

```python
# Tenure in Months 컬럼에서 점수대를 나눔
df['Tenure Group'] = pd.cut(df['Tenure in Months'], bins=[0, 24, 60, 72], labels=['0-24', '24-60', '60+'])
```

```python
plt.figure(figsize=(15,8),dpi=200)
sns.countplot(data=df, x='Tenure Group', hue='CLTV Group', palette='Set2')
```

* 60개월이 넘어가면서 기업/개인 장기 고객들만 남고, 2년 ~ 5년 미만 사이까지 CLTV 낮은 점수대의 고객들이 존재하여 이탈하지 않도록 관리가 필요하다. 

### CLTV 별 이탈율

* 이탈율이 CLTV 4001-6000 대의 이용자들이 잠재적으로 높은 것을 알 수가 있다.
