# 상관분석

## 범주형

### ANOVA (세개 이상의  범주를 가진 범주형 변수)
Membership, Satisfaction Score, Contract

* Membership

```python
# 'Membership' 열을 기준으로 데이터를 그룹화
Membership_groups = [df[df['Membership'] == membership]['Churn Value'] for membership in ['None', 'Offer A', 'Offer B', 'Offer C', 'Offer D', 'Offer E']]

# ANOVA 검정 수행
f_statistic, p_value = stats.f_oneway(*Membership_groups)

print(f"F-statistic: {f_statistic:.4f}")
print(p_value)
print(f"P-value: {p_value:.10f}")
p_value
```

```python
import scipy.stats as stats
import pandas as pd


# ANOVA를 수행합니다.
f_statistic, p_value = stats.f_oneway(
    df['Churn Value'][df['Membership'] == 'None'],# 'Membership' 열이 'None'인 행들의 'Churn Value' 열 값만을 선택
    df['Churn Value'][df['Membership'] == 'Offer A'],
    df['Churn Value'][df['Membership'] == 'Offer B'],
    df['Churn Value'][df['Membership'] == 'Offer C'],
    df['Churn Value'][df['Membership'] == 'Offer D'],
    df['Churn Value'][df['Membership'] == 'Offer E']
)

print("F-statistic:", f_statistic)
print("p-value:", p_value)

# p-value를 통해 검정 결과를 해석합니다.
if p_value < 0.05:  # 일반적으로 사용되는 유의수준 (0.05)을 기준으로 판단합니다.
    print("Membership 그룹 간에 Churn Value에 통계적으로 유의한 차이가 있음")
else:
    print("Membership 그룹 간에 Churn Value에 통계적으로 유의한 차이가 없음")
```

* Satisfaction Score

```python
Membership_groups[0]
```

```python
# 'Satisfaction Score' 열을 기준으로 데이터를 그룹화
SatisfactionScore_groups = [df[df['Satisfaction Score'] == score]['Churn Value'] for score in [1,2,3,4,5]]

# ANOVA 검정 수행
f_statistic, p_value = stats.f_oneway(*SatisfactionScore_groups)

print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.10f}")
p_value
```

### 카이제곱 검정

```python
from scipy.stats import chi2_contingency

# 범주형 변수 리스트
cate_cols = ['Membership', 'Satisfaction Score','Contract' ]

# 결과를 저장할 데이터프레임 초기화
df_cate_corr_label = pd.DataFrame(columns=['Variable', 'Chi-Square Statistic', 'P-value', 'Correlation'])

# for 문을 사용하여 각 범주형 변수와 Churn Label 간의 상관관계를 확인
alpha = 0.05

for col in cate_cols:
    # 크로스 테이블 생성
    cross_table = pd.crosstab(index=df[col], columns=df['Churn Value'])
    
    # 카이제곱 검정 실행
    chi2, p, dof, expected = chi2_contingency(cross_table)
    
    # 결과 해석
    if p < alpha:
        correlation = "유의미한 상관관계"
    else:
        correlation = "유의하지 않은 상관관계"
    

    # 결과를 데이터프레임에 추가
    df_cate_corr_label = df_cate_corr_label.append({'Variable': col, 
                                    'Chi-Square Statistic': chi2, 
                                    'P-value': p, 
                                    'Correlation': correlation}, ignore_index=True)

# 결과 데이터프레임 출력
df_cate_corr_label
```

## 수치형
pointbiserialr 검정 실행 - X변수가 연속형이고, Y변수는 이분변수일 때 사용하는 상관계수

```python
from scipy import stats

# 수치형 변수와 'Churn Label' 열 선택
num_cols_label = ['Age', 'Number of Dependents', 'Tech services', 'Streaming services',
                  'Combined Product','Tenure in Months', 'Monthly Charge', 'Total Revenue']

binary_col = 'Churn Value'

# Point biserial correlation 계산
point_biserial_correlations = {}
for col in num_cols_label:
    correlation, p_value = stats.pointbiserialr(df[col], df[binary_col])
    point_biserial_correlations[col] = correlation

# 결과 출력
print("Point biserial correlations:")
for col, correlation in point_biserial_correlations.items():
    print(f"{col}: {correlation:.2f}")
```

결과

-1.0과 -0.7 사이 -> 강한 음적 선형관계</br>


-0.7과 -0.3 사이 -> 뚜렷한 음적 선형관계</br>
Combined Product, Tenure in Months</br>


-0.3과 -0.1 사이 -> 약한 음적 선형관계</br>
Number of Dependents, Tech services, Total Revenue</br>


-0.1과 +0.1 사이 -> 거의 무시될 수 있는 선형관계</br>
Streaming services</br>


+0.1과 +0.3 사이 -> 약한 양적 선형관계</br>
Age, Monthly Charge</br>


+0.3과 +0.7 사이 -> 뚜렷한 양적 선형관계</br>


+0.7과 +1.0 사이 -> 강한 양적 선형관계

Age: 나이가 많을 수록 이탈율이 높다.</br>
Number of Dependents: 가족결합수가 많을 수록 이탈율이 낮다.</br>
Tech services: 기술서비스 이용개수가 많을수록 이탈율이 낮다.</br>
Streaming services: 부가서비스 이용개수가 많을수록 이탈률이 높다.</br>
Combined Product: 통신 결합상품 이용개수가 많을수록 이탈률이 낮다.</br>
Tenure in Months: 계약기간이 길수록 이탈률이 낮다.</br>
Monthly Charge: 월요금이 높을수록 이탈률이 높다.</br>
Total Revenue: 총요금이 높을수록 이탈률이 낮다.</br>
